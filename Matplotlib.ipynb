```python
x=[1,2,3,4,5]
y1=[3,5,6,8,1]
y2=[1,4,2,6,9]
plt.plot(x,y1,color='r',marker="o",linestyle="--",linewidth=3,markersize=6,label="line y1")
plt.plot(x,y2,color="b",marker="o",linestyle="--",linewidth=5,markersize=10,label="line y2")
plt.title("Compare the plot")
plt.xlabel("dates")
plt.ylabel("profits")
plt.legend()
plt.grid()
plt.show()
```

```python
#stacked histogram >> two histogram together
data1=np.random.normal(0,1,5000)
data2=np.random.normal(3,2,5000)
```

```python
plt.hist([data1,data2],bins=20,color=['r','b'],label=['dist1','dist2'])
plt.xlabel('Values')
plt.ylabel('Frequencies')
plt.title("Stacked hsitogram")
plt.show()
```

```python
# pie chart >. to understand the percent distribution of data
langs=["C","Python","Java"]
students=[20,100,40]
plt.pie(students,labels=langs,autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()
```

```python
#focus on one segment
explode=(0.0,0.0,0.1)
plt.pie(students,labels=langs,autopct="%1.1f%%",explode=explode,shadow=True)
plt.title("Pie Chart")
plt.show()
```

```python
# area plot
# An area plot , also called a stacked area plot ,is a type of data visualization that show how we can show various quantitative factors ahve changed over time while highlighting how they all work together .A layered representation of the data is produced inan area plot by layering many variables on top of it .To distinguish between categories , the space beneath ecah layer is frequently colored.
```

```python
years=[2021,2018,2022,2023,2017]
category1=[20,25,30,28,35]
category2=[10,15,20,18,25]
category3=[5,8,10,9,15]
plt.stackplot(years,category1,category2,category3,labels=["category1","category2","category3"],colors=['orange','skyblue','lightgreen'])
plt.xlabel('years')
plt.ylabel('values')
plt.title("Area plot")
plt.legend()
plt.show()
```
