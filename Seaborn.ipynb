```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

Why seaborn?
1. more attractive plots
2. more options of statistical plots

Seaborn is a Python data visualization library  based on matplotlib . It provides a high-level interface for drawing attractive and informative ststistical graphics.

```python
# numerical data plotting
# relplot has options of plotiing both numerical and categorical data
# relplotis relationship plot (by defualt scatter plot but can make line plot)
```

```python
#  how tip is varying wrt toatal_bill>> scatterplot/by default relplot makes scatter plot
sns.relplot(x='total_bill',y='tip',data=df) # relationship between tow variable
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='smoker') #insight >> as total_bill increases tip increases
#insight>> customer who smoked gave the highest tip
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='size')
```

hue is providing differenet color coding for the dataset

```python
sns.relplot(x='total_bill',y='tip',data=df,style='size') # style gives the different representation of data points
```

```python
sns.relplot(x='total_bill',y='tip',data=df,style='smoker',hue='time')
```

```python
# relplot for making a lineplot >> geenerating a data
#line plot >> to understand the trend in data , a time column/sequence is generally is there
```

```python
time=np.arange(500)
value=np.random.randn(500).cumsum()
df=pd.DataFrame({'time':time,'value':value})
```

```python
sns.relplot(x='time',y='value',kind='line',data=df,sort=True)
```

```python
fmri=sns.load_dataset('fmri')
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,kind='line',hue='event',style='region',dashes=False)# remove dashes from line chart
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,kind='line',hue='event',style='region',markers=True)
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
# two subplots based on a column
sns.relplot(x='total_bill',y='tip',hue='smoker',data=tips,col='time')
```

```python
# categorical data plots
# for categorical data we have catplot
sns.catplot(x='day',y='total_bill',data=tips)# plot between day and total bill
```

insight many people arrive on weekne then on week days

```python
#in catplot ,few datapoints are overlapping>> to see it separate use swarm plot
sns.catplot(x='day',y='total_bill',data=tips,kind="swarm")
```

```python
#boxplot  >> to see outliers
sns.catplot(x='day',y='total_bill',data=tips,kind='box')
```

```python
#violin plot>>
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```

```python
#boxen plot is made from box plot itself
sns.catplot(x='day',y='total_bill',data=tips,kind='boxen')
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
#how total bill(aggregrated) varies with different day
```

```python
sns.jointplot(x=tips.total_bill,y=tips.tip) #draw a plot of two variables  woth distribution>>bivariate plots
```

```python
# joint plot>> understand relationshipm and also distribution of two variables
sns.jointplot(x=tips.total_bill,y=tips.tip,kind="reg") # reg>>regression line to represent reltaionship between
```

```python
sns.jointplot(x=tips.total_bill,y=tips.tip,kind='hex')
```

```python
sns.jointplot(x=tips.total_bill,y=tips.tip,kind='kde') # kernel density estimator,contour>>meterological department
```

```python
sns.pairplot(tips) #make plot scatter plot with respect to all numerical columns
```

```python
sns.pairplot(tips,hue="smoker")
```

```python
sns.pairplot(tips,hue="smoker",palette="coolwarm")
```

```python
# boxplot using sns directly(earlier catplot)'
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.boxplot(data=tips,palette='rainbow',orient='h')
```
