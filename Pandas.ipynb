```python
# merging/joining Two dataframe using vein diagram
# >>intersection(inner merge)
# >>part A(merge based on left dataframe is called left merge(All the elements of A datafrmae will be in the merge))
# >>also similarly there is a right data frame
# >>Outermerge (all the elements of both dataframes)
# >>cross merge (every permutation of rows)
```

```python
#merge
df1=pd.DataFrame({
    "key1":[1,2,4,5,6],
    "key2":[4,5,6,7,8],
    "key3":[3,4,5,6,6]
})
df1
```

```python
df2=pd.DataFrame({
    "key1":[1,2,45,6,67],
    "key4":[56,5,6,7,8],
    "key5":[3,56,5,6,6]
})
df2
```

```python
pd.merge(df1,df2,how="inner") #
```

```python
pd.merge(df1,df2,how="left")#
```

```python
pd.merge(df1,df2,how="right")
```

```python
#Common elemnets are known as keys
pd.merge(df1,df2,how="outer") #all the common elements of the common column from both the dataframe
```

```python
pd.merge(df1,df2,how="cross") # every permutation of rows
```

```python
pd.merge(df1,df2,how="left",left_on="key2",right_on="key4") #
```

```python
#join >.join works based on index/merge works based on columns
df1=pd.DataFrame({
    "key1":[1,2,4,5,6],
    "key2":[4,5,6,7,8],
    "key3":[3,4,5,6,6]
},index=["a","b","c","d","e"])
df1
```

```python
df2=pd.DataFrame({
    "key6":[1,2,45,6,67],
    "key4":[56,5,6,7,8],
    "key5":[3,56,5,6,6]
},index=["a","b","h","i","j"])
df2
```

```python
df1.join(df2,how="inner") # joins columns of another dataframe based on common indexes
```

```python
df1.join(df2,how="left") # joins columns of another dataframe based on common indexes
```

```python
df1.join(df2,how="right") # joins columns of another dataframe based on common indexes
```

```python
df1.join(df2,how="outer") # joins columns of another dataframe based on common indexes
```

```python
df1.join(df2,how="cross")# all permutation
```
