```python
#implicit indexing
df.iloc[0:2]# start from 0 and go till 1
```

```python
#explicit indexing
df.loc[0:3]# start from 0 and go till 3
# loc is used fornamed indexes, iloc is used for integer indexes
```

```python
# df.iloc[0:2,["Name","Sex"]] # this will throw an error because iloc used default integer indexs(the one which we use when there is header="None")
# so instead of using Name and Sex use intgere indexes
df.iloc[0:10,[1,2]]
```

```python
# and for using named indexes we can use loc
df.loc[0:3,["Name","Sex"]]
```
